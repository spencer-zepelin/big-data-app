# Weather EDA

```python
import pandas as pd
```

```python
df = pd.read_csv('final_proj_weather_data.csv')
```

```python
df.head()
```

```python
df = df.drop(0)
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.astype({'DATE': 'datetime64', 'AWND':'float32', 'FMTM':'float32', 'PGTM':'float32','PRCP':'float32','TMAX':'float32','TMIN':'float32'})
```

```python
df.dtypes
```

```python
df = df.fillna(df.median())
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["AvgWindSpeed-MPH"] = (df["AWND"] / 10) * 2.23694
```

```python
df["Precipitation-IN"] = (df["PRCP"]/10) /25.4
```

```python
df["MaxTemp-F"] = (df["TMAX"] / 10) * 1.8 + 32
df["MinTemp-F"] = (df["TMIN"] / 10) * 1.8 + 32
```

```python
df.head()
```

```python
df.describe()
```

```python
df["WindBins"] = df.apply(lambda x: 'L' if x['AvgWindSpeed-MPH'] < 7.2 else ('M' if x['AvgWindSpeed-MPH'] < 11.9 else 'H'), axis=1)
df["HighTempBins"] = df.apply(lambda x: 'L' if x['MaxTemp-F'] < 43 else ('M' if x['MaxTemp-F'] < 79 else 'H'), axis=1)
df["PrecipitationBins"] = df.apply(lambda x: 'L' if x['Precipitation-IN'] < 0.04 else "H", axis=1)
```
